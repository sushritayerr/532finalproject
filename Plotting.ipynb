```python
import matplotlib.pyplot as plt

def plot(x_values, y_values, x_label, y_label, title):
    plt.plot(x_values, y_values, marker='o', linestyle='-')
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.title(title)

    plt.show()
```
